### Concatenating Strings and Integers

```python
my_string = "The answer is "
my_integer = 42

# This will cause a TypeError:
# try:
#     result_error = my_string + my_integer
# except TypeError as e:
#     print(f"TypeError: {e}")

# Correct way 1: Convert integer to string using str()
result_str_func = my_string + str(my_integer)
print(f"Using str() function: {result_str_func}")

# Correct way 2: Using an f-string (preferred method for readability)
result_fstring = f"{my_string}{my_integer}"
print(f"Using f-string: {result_fstring}")
```

```python
# Conditional Statements: Allow programs to execute blocks of code based on certain conditions
# if, Elif, else
# Example : Loan Calculator
```

```python
name = "Temilara"
age = 28
salary = 200000
loan_amount = 20000000
interest = loan_amount * 0.5
```

```python
if age >= 18 and (salary * 12) > (loan_amount + interest): #Indentation added (:)
  print (f"{name} can take a loan")
else:
  print (f"{name} cannot take a loan")
```
